```python
import cv2
import numpy as np
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
cv2.imshow("Output",img)                            #to display image we use imshow()
cv2.waitKey(1)                                      #waitKey is attribute
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)       #it(cv2.COLOR_) has different color.
#mgGray = cv2.cvtColor(img,cv2.COLOR_BAYER_BG2BGRA)
cv2.imshow("Gray Image",imgGray)                        #it shows gray image
imgBlur = cv2.GaussianBlur(imgGray,(7,7),0)
cv2.imshow("Blur Image",imgBlur)                        #it shows Blur image
cv2.waitKey(2)                                      
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
imgCanny = cv2.Canny(img,100,100)                   #thrshold value
cv2.imshow("Canny Image",imgCanny)                  #to find the edges of the image #edges are low by increase value.
cv2.waitKey(1) 
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
kernel=np.ones((5,5),np.uint8)                             #we want all value are 1 and unit8 gives range from 0-255
imgDialation =cv2.dilate(imgCanny,kernel,iterations=1)
cv2.imshow("Dialation Image",imgDialation)               #to dectect proper line we can increase thikness of edges we use dialation.
cv2.waitKey(1)                                           #kernal is just a matrix defined by 1
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
kernel=np.ones((5,5),np.uint8)                              
imgEroded=cv2.erode(imgDialation,kernel,iterations=1)    #opposite of dialation to get thiner edges we use eroded function.
cv2.imshow("Eroded Image",imgEroded)
cv2.waitKey(1000)
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
kernel=np.ones((5,5),np.uint8)
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imgBlur = cv2.GaussianBlur(imgGray,(7,7),0)
imgCanny = cv2.Canny(img,100,100)
imgDialation =cv2.dilate(imgCanny,kernel,iterations=1)
imgEroded=cv2.erode(imgDialation,kernel,iterations=1)

cv2.imshow("Gray Image",imgGray)
cv2.imshow("Blur Image",imgBlur)
cv2.imshow("Canny Image",imgCanny)
cv2.imshow("Dialation Image",imgDialation)

cv2.imshow("Dialation Image",imgDialation)
cv2.imshow("Eroded Image",imgEroded)
cv2.waitKey(1)
```

## RESIZING AND CROPPING-

```python
#original image
img=cv2.imread("C:/Users/com/Desktop/AIS/lena.jpg")
print(img.shape)                                      #it gives shape of height width and the channal 3 for BGR
#RESIZING
imgResize=cv2.resize(img,(200,500))      #by using resize function we can change the size of image (width,height)
print(imgResize.shape)
#CROPPING
imgCropped=img[200:500,200:500]            #we used functionality not cv2 for cropped the image.(height range,width range)
cv2.imshow("Image",img)
cv2.imshow("Resize Image",imgResize) 
cv2.imshow("Cropped Image",imgCropped)
cv2.waitKey(1)
```

```python
img=np.zeros((512,512,3),np.uint8)
cv2.line(img,(0,0),(img.shape[1],img.shape[0]),(0,255,0),3)            #shape[1]=width,shape[0]=height
cv2.rectangle(img,(0,0),(250,350),(0,0,255),2)                           #3rd is for color and last for thickness
cv2.rectangle(img,(0,0),(250,350),(0,0,255),cv2.FILLED)
cv2.circle(img,(400,50),30,(255,255,0),5)                       #1.center point,2.radius,3.color,4.thickness
cv2.putText(img,"OPENCV",(300,200),cv2.FONT_HERSHEY_COMPLEX,1,(0,150,0),4) 
                                            #1.text,2.origin,3.font of text,4.scale(size),5.color,6.thickness of text
cv2.imshow("Image",img)  
cv2.waitKey(1)
#cv2 has different font in the library.
```

## WARP PERSPECTIVE

```python
import cv2
import numpy as np
```

```python
img=cv2.imread("C:/Users/com/Desktop/AIS/cards.jpg")
width,height = 250,350
pts1=np.float32([[111,219],[287,188],[154,482],[352,440]])            #4 corner points 
pts2=np.float32([[0,0],[width,0],[0,height],[width,height]])
matrix=cv2.getPerspectiveTransform(pts1,pts2)
imgOutput=cv2.warpPerspective(img,matrix,(width,height))
cv2.imshow("Output",img)                                      #to display image we use imshow()
cv2.imshow("Output image",imgOutput)
cv2.waitKey(1)
```
